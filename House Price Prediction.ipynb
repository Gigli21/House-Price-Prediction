**Handling Missing Data**

```python
#Check percentage of nan values in each feature
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1]


for feature in features_with_na:
    print(feature, np.round(dataset[feature].isnull().mean(), 4), '% missing values')
```

```python
#Check relationship between missing values and Sales Price

for feature in features_with_na:
    data = dataset.copy()

#Create a variable that indicates 1 if there are nan values and 0 if not nan
data[feature] = np.where(data[feature].isnull(), 1, 0)

#Calculate the mean SalesPrice where the information is missing or present
data.groupby(feature)['SalePrice'].median().plot.bar()
plt.title(feature)
plt.show()
```

Plot above shows that the price is affected by the nan values as the price is highest with high number of values

**Dealing with numerical variables**

```python
#Check number of numerical variables
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes !='O']
print('Number of numerical variables: ', len(numerical_features))
```

```python
#Visualise the numerical variables
dataset[numerical_features].head
```

**Dealing with temporal variables**

```python
#Creating a list of features that contain year information
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]

year_feature
```

```python
#View the unique variables of year_feature
for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
#Check whether there is any correlation between year sold and sale price

dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title("House Price vs Year Sold")
```

```python
#Compare the difference between all variables in year_feature with YrSold

for feature in year_feature:
    if feature !='YrSold':
        data = dataset.copy()
        data[feature] = data['YrSold'] - data[feature]
        
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
#Checking how many discrete features in dataset
discrete_feature = [feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
print("Discrete Variables Count: {}".format(len(discrete_feature)))
```
